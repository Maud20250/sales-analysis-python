# Sales Analysis Project
Analyse des ventes par produit, categorie et region.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/sales_data.csv')
df['total_sales'] = df['quantity'] * df['unit_price']
print(df.head())
print(df.shape)
```

```python
category_sales = df.groupby('category')['total_sales'].sum()
print(category_sales)
```

```python
category_sales.plot(kind='bar', color=['steelblue','coral'])
plt.title('Total Sales by Category')
plt.tight_layout()
plt.savefig('../images/sales_by_category.png')
plt.show()
```

```python
region_sales = df.groupby('region')['total_sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', color='teal')
plt.title('Total Sales by Region')
plt.tight_layout()
plt.savefig('../images/sales_by_region.png')
plt.show()
```
